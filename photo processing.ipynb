# Color Transformation

```python
from skimage import color
```

```python
image=io.imread("Birds.jpg",as_gray=True)
image
```

```python
image.shape
```

```python
io.imshow(image)
```

```python
io.imshow(image)
```

```python
gray=color.rgb2gray(image)
io.imshow(gray)
```

```python
hsv=color.rgb2hsv(image)
io.imshow(hsv,cmap='hsv')
```

```python
rgb=color.hsv2rgb(image)
io.imshow(rgb)
```

# Quality Transformation

```python
from skimage import transform
```

```python
io.imshow(image)
```

```python
rescale_image=transform.rescale(image,0.5,multichannel=True )
io.imshow(rescale_image)
```

```python
cropped=transform.resize(image,output_shape=(128,128))
io.imshow(cropped)
```

```python
rotated=transform.rotate(image,angle=90)
io.imshow(rotated)
```

```python
rotated=transform.rotate(image,angle=90,resize=True)
io.imshow(rotated)
```

# Exposure

```python
from skimage import exposure
```

```python
gamma=exposure.adjust_gamma(image,0.8)
io.imshow(gamma)
```

```python
log=exposure.adjust_log(image,1.5)
io.imshow(log)
```

```python
sigmoid=exposure.adjust_sigmoid(image,0.5)
io.imshow(sigmoid)
```
